```python
import numpy as np
import matplotlib.pyplot as plt

# 100개의 난수 생성 (0~1 사이)
data_set = np.random.rand(100)  # 변수를 먼저 정의!

# 히스토그램 시각화
plt.hist(data_set, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# 기본 통계값 출력
print("최댓값:", np.max(data_set))
print("최솟값:", np.min(data_set))
print("평균값:", np.mean(data_set))
print("중앙값:", np.median(data_set))
```

```python
import numpy as np

b = [[5, 3, 4], [2, 5, 1], [7, 6, 9]]  # 2D 리스트
arr_b = np.array(b)  # NumPy 배열로 변환

print("🔹 numpy 배열 변환 결과:")
print(arr_b)

print("\n🔹 numpy 정렬1: np.sort(arr_b) (기본 정렬 - 행 기준)")
print(np.sort(arr_b))  # 기본적으로 행 기준 정렬 (axis=1과 동일)

print("\n🔹 numpy 정렬2: np.sort(arr_b, axis=1) (행 기준 오름차순 정렬)")
print(np.sort(arr_b, axis=1))  # 행별 정렬

print("\n🔹 numpy 정렬3: np.sort(arr_b, axis=0) (열 기준 오름차순 정렬)")
print(np.sort(arr_b, axis=0))  # 열별 정렬

print("\n🔹 numpy 정렬4: np.sort(arr_b, axis=0)[::-1] (열 기준 내림차순 정렬)")
print(np.sort(arr_b, axis=0)[::-1])  # 열 기준 내림차순 정렬
```

```python
# 🔹 기본 실습 1: Numpy 배열 생성 및 슬라이싱
# 문제:
#
# 2차원 Numpy 배열 3x3 을 생성하고,
# 특정 행과 열을 슬라이싱하여 출력하세요.
# - 첫 번째 행 가져오기
# - 2번째 열만 가져오기
# - 1행 1열~끝까지
import numpy as np

# 3x3 난수 배열 생성
np_arr = np.random.randn(3, 3)

print("전체 배열:")
print(np_arr)

# 첫 번째 행 가져오기
print("\n첫 번째 행:")
print(np_arr[0, :])

# 두 번째 열 가져오기
print("\n두 번째 열:")
print(np_arr[:, 1])

# (1,1)부터 끝까지 슬라이싱
print("\n1행 1열~끝까지:")
print(np_arr[1:, 1:])
```

```python
# 🔹 기본 실습 1: Numpy 배열 생성 및 슬라이싱
# 문제:
#
# 2차원 Numpy 배열 3x3 을 생성하고,
# 특정 행과 열을 슬라이싱하여 출력하세요.
# - 첫 번째 행 가져오기
# - 2번째 열만 가져오기
# - 1행 1열~끝까지

import numpy as np

# 난수 시드 설정 (실행할 때마다 같은 결과)
np.random.seed(42)

# 1~10 범위의 정수로 이루어진 3x3 배열 생성
np_arr = np.random.randint(1, 11, (3, 3))

print("전체 배열:")
print(np_arr)

# 첫 번째 행 가져오기
print("\n첫 번째 행:")
print(np_arr[0, :])

# 두 번째 열 가져오기
print("\n두 번째 열:")
print(np_arr[:, 1])

# (1,1)부터 끝까지 슬라이싱
print("\n1행 1열~끝까지:")
print(np_arr[1:, 1:])
```

```python
# 🔹 기본 실습 2: 배열 연산
# 문제:
# 두 개의 Numpy 배열을 생성하고,
# 더하기(+), 빼기(-), 곱하기(*), 나누기(/) 연산을 수행하세요.
# [1, 2, 3]
# [4, 5, 6]

import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])
print(f"덧셈: {a+b}")
print(f"빼기: {a-b}")
print(f"곱하기: {a*b}")
print(f"나누기: {a/b}")
```

```python
# 🔹 기본 실습 2: 배열 연산
# 문제:
# 두 개의 Numpy 배열을 생성하고,
# 더하기(+), 빼기(-), 곱하기(*), 나누기(/) 연산을 수행하세요.
# [1, 2, 3]
# [4, 5, 6]

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 덧셈
print(f"덧셈: {a + b}")

# 빼기
print(f"빼기: {a - b}")

# 곱하기
print(f"곱하기: {a * b}")

# 나누기 (0으로 나누는 경우 처리)
# numpy에서는 0으로 나누는 경우 'inf'로 표시되거나 'NaN'이 나올 수 있기 때문에 예외 처리
try:
    print(f"나누기: {a / b}")
except ZeroDivisionError:
    print("0으로 나누는 오류 발생!")
```
