```python
import numpy as np

b = [[5, 3, 4], [2, 5, 1], [7, 6, 9]]  # 2D 
arr_b = np.array(b)  # NumPy  

print(" numpy   :")
print(arr_b)

print("\n numpy 1: np.sort(arr_b) (  -  )")
print(np.sort(arr_b))  #     (axis=1 )

print("\n numpy 2: np.sort(arr_b, axis=1) (   )")
print(np.sort(arr_b, axis=1))  #  

print("\n numpy 3: np.sort(arr_b, axis=0) (   )")
print(np.sort(arr_b, axis=0))  #  

print("\n numpy 4: np.sort(arr_b, axis=0)[::-1] (   )")
print(np.sort(arr_b, axis=0)[::-1])  #    
```

```python
#    1: Numpy    
# :
#
# 2 Numpy  3x3  ,
#     .
# -    
# - 2  
# - 1 1~
import numpy as np

# 3x3   
np_arr = np.random.randn(3, 3)

print(" :")
print(np_arr)

#    
print("\n  :")
print(np_arr[0, :])

#    
print("\n  :")
print(np_arr[:, 1])

# (1,1)  
print("\n1 1~:")
print(np_arr[1:, 1:])
```

```python
#    1: Numpy    
# :
#
# 2 Numpy  3x3  ,
#     .
# -    
# - 2  
# - 1 1~

import numpy as np

#    (   )
np.random.seed(42)

# 1~10    3x3  
np_arr = np.random.randint(1, 11, (3, 3))

print(" :")
print(np_arr)

#    
print("\n  :")
print(np_arr[0, :])

#    
print("\n  :")
print(np_arr[:, 1])

# (1,1)  
print("\n1 1~:")
print(np_arr[1:, 1:])
```

```python
#    2:  
# :
#   Numpy  ,
# (+), (-), (*), (/)  .
# [1, 2, 3]
# [4, 5, 6]

import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])
print(f": {a+b}")
print(f": {a-b}")
print(f": {a*b}")
print(f": {a/b}")
```

```python
#    2:  
# :
#   Numpy  ,
# (+), (-), (*), (/)  .
# [1, 2, 3]
# [4, 5, 6]

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 
print(f": {a + b}")

# 
print(f": {a - b}")

# 
print(f": {a * b}")

#  (0   )
# numpy 0   'inf'  'NaN'      
try:
    print(f": {a / b}")
except ZeroDivisionError:
    print("0   !")
```
