```python
import numpy as np
import pandas as pd
import matplotlib

#50개의난수생성
data=np.random.rand(50)
print(type(data))
print(data)

# 넘파이 배열을 판다스의 시리즈 자료형으로 변환(인덱스,데이터의 조합)
seri=pd.Series(data)
print(type(seri))
print(seri)
seri.plot()
seri.plot.bar()
```

```python
y1=np.ones(10) #1차원 배열에 1채우기
y2=np.ones([3,4]) #2차원 배열에 1채우기
y3=np.ones(12).reshape(3,4) # y2와 동일
print("y1")
print(y1)
print("y2")
print(y2)
print("y3")
print(y3)
```

```python
y4=np.zeros(10) #1차원 배열에 1채우기
y5=np.zeros((3,4)) #2차원 배열에 1채우기
y6=np.zeros(12).reshape(3,4) # y5와 동일
print("y4")
print(y4)
print("y5")
print(y5)
print("y6")
print(y6)
```
